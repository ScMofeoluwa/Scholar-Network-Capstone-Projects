#**Data Preparation**

```
import numpy as np
import pandas as pd
```

```
pd.set_option('display.max_rows',None)
pd.set_option('display.max_columns',None)
```

```
train_df = pd.read_csv('train.csv')
test_df = pd.read_csv('test.csv')
ss = pd.read_csv('SampleSubmission.csv')
des = pd.read_csv('VariableDescription.csv')
```

```
des
```

```
train_df.head()
```

```
train_df.isnull().sum()
```

```
s = train_df.isnull().sum()
res = []
for i,j in s.items():
  if j > 6042:
    res.append(i)
print(len(res))
```

**Drop columns that weren't highly correlated to the target**

```
train_df.drop(res,axis=1,inplace=True)
test_df.drop(res,axis=1,inplace=True)
```

```
train_df.head()
```

```
train_df.isnull().sum()
```

```
train_df['CTR_CATEGO_X'].value_counts()
```

**Fill missing values with 0**

```
train_df.fillna(value=0,inplace=True)
```

```
train_df.isnull().sum()
```

```
train_df.head()
```

**Handle categorical values**

```
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
```

```
train_df['CTR_CATEGO_X'] = LE.fit_transform(train_df[['CTR_CATEGO_X']])
```

```
train_df.head()
```

```
test_df.fillna(value=0,inplace=True)
```

```
test_df['CTR_CATEGO_X'] = LE.transform(test_df[['CTR_CATEGO_X']])
```

#**Training our model**

```
X = train_df.drop(['id','target'],axis=1)
y = train_df['target']
```

```
from sklearn.model_selection import train_test_split
```

```
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=1,test_size=0.3)
```

```
!pip install catboost
```

```
from catboost import CatBoostRegressor
```

```
cat = CatBoostRegressor(eval_metric='RMSE',random_state=1,random_strength=0,n_estimators=5000,
                        learning_rate=0.02,depth=12,l2_leaf_reg=10,
                        bagging_temperature=0.2,od_type='Iter',metric_period=75,od_wait=100)
```

```
cat.fit(X_train,y_train,
        eval_set=[(X_train,y_train),(X_test,y_test)],early_stopping_rounds=500,
        use_best_model=True,verbose=True)
```

```
cat.fit(X,y)
```

```
from sklearn.metrics import mean_squared_error as mse
```

```
pred = cat.predict(X_test)
np.sqrt(mse(y_test,pred))
```

```
tests = test_df.drop('id',axis=1)
```

```
test_pred = cat.predict(tests)
```

```
ss.head()
```

```
ss.target = test_pred
```
