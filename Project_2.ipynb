#**Handling data imbalanced data**

```
#After making both submissions, I noticed the dataset was imbalanced
#It had a ratio of 1:24 so I have to handle imbalance by oversampling and undersampling
```

```
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from imblearn.pipeline import Pipeline
```

```
over = SMOTE(sampling_strategy=0.1)
under = RandomUnderSampler(sampling_strategy=0.5)
steps = [('o', over), ('u', under)]
pipeline = Pipeline(steps=steps)
```

```
X, y = pipeline.fit_resample(X, y)
```

```
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=101)
```

#**Cross-Validation**

```
from sklearn.model_selection import KFold
```

```
kf = KFold(n_splits=5,random_state=43,shuffle=True)
scores = []
test_scores = np.zeros(len(test_df))
for fold,(tr_in,te_in) in enumerate(kf.split(X)):
    print("================================Fold{fold}====================================".format(fold=fold))
    X_train,X_test = X[tr_in],X[te_in]
    y_train,y_test = y[tr_in],y[te_in]
    lgb = LGBMClassifier(random_state=42)
    lgb.fit(X_train,y_train,eval_set=[(X_train,y_train),(X_test,y_test)],early_stopping_rounds=500,verbose=1000)
    scores.append(accuracy_score(y_test,lgb.predict(X_test)))
    test_scores += lgb.predict(test_df)
```

```
np.mean(scores)
```

```
test_pred = test_scores/5
ss['TARGET'] = test_pred
```

```
ss.to_csv('submission_lgb2.csv',index=False)
```

```
#accuracy of 0.72600
```
